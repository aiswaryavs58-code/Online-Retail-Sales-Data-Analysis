# **Exploratory Data Analysis (EDA)**

EDA means carefully looking at the data to understand it better before making any conclusions.

When we collect data in real life, it is usually just a large collection of numbers and text. By simply looking at the data, it is hard to understand:


*   What is happening in the data
*   Which values are important



*   Whether there are any patterns or problems


EDA helps us:


*   Understand the data clearly
*   Find patterns and trends





*   Compare values



*   Identify unusual or incorrect data

Imagine you receive a long list of monthly expenses.
EDA is like checking how much you usually spend, where you spend the most, and whether any month has unusually high spending.

# **Data Visualization**

Data visualization means showing data in the form of pictures like charts and graphs.

Numbers written in tables can be confusing, especially for large datasets.
Graphs make information:


*   Easy to understand
*   Easy to explain




*   Easy to remember


Visualization helps people quickly see:


*   Which values are high or low
*   How data changes over time

*   Relationships between different values


Simple example:
Instead of reading 100 numbers about sales, a single bar chart can clearly show which product sells the most.

### **Why EDA and Visualization Are Important**


*   They help us understand the data before making decisions

*   They reduce chances of wrong conclusions
*   They make data easy to explain to non-technical people


*   They help identify patterns, trends, and problems


### **How EDA and Visualization Work Together**


*   EDA helps us analyze and summarize the data
*   Visualization helps us present those findings clearly

Together, they turn raw data into meaningful information.

# **Univariate Analysis (One Column at a Time)**

Univariate analysis means studying one type of information at a time to understand it clearly. In this method, we focus on a single column or value and observe how it behaves. For example, if we look only at the quantity of products sold, we can understand what quantities are common, which ones are very high or very low, and what the average value is. This helps us get a basic idea of the data without comparing it to anything else. Univariate analysis is mainly used to understand the overall nature of one variable.

**1. Distribution of Quantity Purchased**

```python
plt.figure(figsize=(6,4))
plt.hist(df['Quantity'], bins=30)
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.title("Distribution of Quantity Purchased")
plt.show()
```

This graph shows how many items customers usually buy in a single purchase.


*   The horizontal line (X-axis) shows the number of items bought (Quantity).

*   The vertical line (Y-axis) shows how many times that quantity appears (Frequency).

*   Each bar represents how often a certain quantity is purchased.


In simple words, this chart helps us understand:


*   What quantity is most commonly purchased
*   Whether customers usually buy small or large quantities


*   If there are rare cases where very high quantities are bought


**2. Distribution of Unit Price**

# **Bivariate Analysis (Comparing Two Columns)**

Bivariate analysis means studying two types of information together to see how they are related. Instead of looking at values separately, we compare them. For example, if we look at quantity sold and total price together, we can understand whether selling more items leads to higher sales. This type of analysis helps us identify relationships and patterns between two variables. It answers questions like â€œDoes one value affect the other?â€

**6. Quantity vs TotalPrice (Scatter Plot)**

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Quantity'], df['TotalPrice'])
plt.xlabel("Quantity")
plt.ylabel("Total Price")
plt.title("Quantity vs Total Price")
plt.show()
```

This chart shows the relationship between how many items are bought and how much money is spent in total.

On this graph, the horizontal line shows the quantity of items purchased, and the vertical line shows the total price paid. Each dot represents one purchase. When dots move upward as they move to the right, it means that buying more items usually leads to paying more money.

In simple terms, this visual helps us understand whether customers who buy more items also spend more. If most dots follow an upward pattern, it shows a clear relationship between quantity and total price. This chart makes it easy to see buying behavior without reading individual bills or numbers.

**7. Total Sales by Country**

```python
sales_by_country = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(7,4))
sales_by_country.plot(kind='bar')
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.title("Total Sales by Country")
plt.show()
```

This chart shows **which countries generate the highest total sales**.

The **countries are listed along the bottom**, and the **height of each bar shows the total amount of sales from that country**. All individual purchases from each country are added together to get this total. Taller bars mean **higher sales**, while shorter bars mean **lower sales**.

In simple terms, this visual helps us understand **which countries bring in the most revenue**. It allows us to quickly identify the **top-performing countries** without checking thousands of individual sales records, making it easy to see where the business is strongest.


**8. Average TotalPrice by Payment Method**

```python
avg_payment = df.groupby('PaymentMethod')['TotalPrice'].mean()

plt.figure(figsize=(6,4))
avg_payment.plot(kind='bar')
plt.xlabel("Payment Method")
plt.ylabel("Average Total Price")
plt.title("Average Total Price by Payment Method")
plt.show()
```

This chart shows **the average amount customers spend based on the payment method they use**.

Each bar represents a **different payment method** such as card, cash, or online payment. The **height of the bar shows the average total price** spent by customers using that method. Instead of adding all sales, this chart looks at the **typical spending amount** for each payment type.

In simple words, this visual helps us understand **which payment methods are usually linked to higher or lower spending**. For example, if one bar is taller, it means customers using that payment method **tend to spend more on average**. This makes it easy to compare customer spending behavior across different payment options.


**9. Return Status vs TotalPrice**

```python
return_sales = df.groupby('ReturnStatus')['TotalPrice'].sum()

plt.figure(figsize=(5,4))
return_sales.plot(kind='bar')
plt.xlabel("Return Status")
plt.ylabel("Total Sales")
plt.title("Return Status vs Total Price")
plt.show()
```

This chart shows how total sales are affected by whether products are returned or not.

The horizontal axis shows the return status, such as items that were returned and items that were not returned. The height of each bar shows the total sales amount for each return category, calculated by adding up all purchases under that status.

In simple terms, this visual helps us understand how much money comes from completed sales compared to returned orders. A taller bar for non-returned items means most sales are successfully completed, while a smaller bar for returned items shows the portion of sales that did not remain final. This makes it easy to see the impact of returns on overall sales.

**10. Discount vs TotalPrice**

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Discount'], df['TotalPrice'])
plt.xlabel("Discount")
plt.ylabel("Total Price")
plt.title("Discount vs Total Price")
plt.show()
```

This chart shows the **relationship between discounts and the final amount customers spend**. Each dot represents one purchase, where the **discount given** is shown along the bottom and the **total price paid** is shown on the side. By looking at how the dots are spread, we can understand how discounts affect spending. Generally, higher discounts often lead to lower total prices, but in some cases customers may still spend more if they buy more items. Overall, this visual helps us see **whether giving discounts actually reduces the bill amount or encourages customers to purchase more**, in a very easy and visual way.


# **Multivariate Analysis (Multiple Columns Together)**

Multivariate analysis means studying more than two types of information at the same time. This is done to understand complex patterns that cannot be seen when looking at one or two values alone. For example, analyzing quantity, price, country, and total sales together helps us understand how different factors work together. Multivariate analysis gives a deeper and more complete understanding of the data by considering multiple aspects at once.

**11. Correlation Heatmap (Numeric Features)**

```python
import seaborn as sns

numeric_df = df[['Quantity', 'UnitPrice', 'Discount', 'TotalPrice']]
correlation = numeric_df.corr()

plt.figure(figsize=(6,4))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

This chart is called a **correlation heatmap**, and it helps us understand **how different numbers are related to each other**. Each box shows how strongly two things move together, such as quantity, unit price, discount, and total price. **Warm colors (like red)** mean a strong relationship â€” when one value increases, the other usually increases too. **Cool colors (like blue)** mean the opposite â€” when one goes up, the other tends to go down. Values close to **0** mean there is little or no relationship. For example, if *Quantity* and *Total Price* are strongly positive, it means buying more items usually increases the total bill. Overall, this visual makes it easy to see **which factors influence each other the most**, without needing technical knowledge.


**12. Total Sales by Country & Sales Channel**

```python
country_channel_sales = df.groupby(['Country', 'SalesChannel'])['TotalPrice'].sum().unstack().head(10)

country_channel_sales.plot(kind='bar', figsize=(8,4))
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.title("Total Sales by Country and Sales Channel")
plt.show()
```

This chart shows **how much total sales come from each country, split by sales channel (online and in-store)**. For every country on the bottom, you can see **two bars side by side**, each one representing a different way customers bought the products. By comparing the heights of these bars, we can easily understand **which country sells more overall** and also **whether people in that country prefer shopping online or in physical stores**. For example, if the online bar is taller than the in-store bar for a country, it means most sales there happen online. This visual helps businesses see **country-wise buying habits** and decide where to focus online or in-store efforts.


**13. Total sales by Country and InvoiceDate (monthly)**

```python
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')

pivot_sales = pd.pivot_table(
    df,
    values='TotalPrice',
    index='Country',
    columns='InvoiceMonth',
    aggfunc='sum'
)

pivot_sales.head()
```

This table shows **how total sales change over time for each country, month by month**. Each row represents a **country**, and each column represents a **month**, while the numbers inside the table show **how much money was earned in that country during that month**. By looking at this, even a non-technical person can easily understand **which months had higher or lower sales in different countries**. It helps answer simple questions like *â€œIn which month did a country perform best?â€* or *â€œAre sales increasing or decreasing over time?â€*. Overall, this view helps businesses track **monthly sales trends by country** and plan better for future months.


**14. Monthly Sales Trend**

```python
monthly_sales = df.groupby('Month')['TotalPrice'].sum()

plt.figure(figsize=(7,4))
plt.plot(monthly_sales)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

This line chart shows **how total sales change from month to month** over a period of time. Each point on the line represents the **total amount of sales in a particular month**, and the line connects these points so we can clearly see the **overall pattern**. If the line goes up, it means sales are increasing; if it goes down, it means sales are decreasing. This visual makes it very easy for a non-IT person to understand **which months performed well, which months were slow, and whether the business is generally growing or declining over time**. It helps businesses plan ahead by identifying **busy seasons and low-sales periods**.
